# Main project

```python


from bs4 import BeautifulSoup
import requests
import time

def job_scraper(skills_known):
        page=requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=')
        soup=BeautifulSoup(page.text,'lxml')
        jobs=soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
        count=0
        for index,job in enumerate(jobs):
            comp=job.find('h3',class_='joblist-comp-name')
            comp_str=comp.text
            skills=job.find('span',class_='srp-skills')
            skills_str=skills.text
            date=job.find('span', class_='sim-posted')
            date_str=date.text.strip()
            link=job.header.h2.a['href']
            #print(comp_str)
            if skills_known in skills_str:
                if "few" in date_str:
                    count=1
                    with open(f'posts/{index}.txt','w') as f:
                        f.write(f" Company name: {comp_str.strip()}\n") 
                        f.write(f" Required Skills : {skills_str.strip().replace(' ','')}\n ")
                        f.write(f"More info : {link} \n")
                    print(f'''
        Company name: {comp_str.strip()} 
        Required Skills : {skills_str.strip().replace(' ','')} 
        More info : {link} \n\n ''')
        if count==0:
            print("\n No Jobs found with specified keyword") 
        
skills_known=input("Enter a skill / keyword to search:")
print(f'''Jobs with '{skills_known}' are being searched ... ''')
job_scraper(skills_known)
        
if __name__=='__main__':
     while True:
          job_scraper(skills_known)
          time_to_wait=0.1
          print(f'waiting for {time_to_wait} minutes ...')
          time.sleep(time_to_wait*60)

```
